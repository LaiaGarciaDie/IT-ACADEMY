```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = r'G:\Mi unidad\ESTUDIS\1.TECH\IT ACADEMY\1. DATA ANALISIS\6. FINAL\ObesityDataSet - transf.xlsx'
df = pd.read_excel(file_path)
```

```python
df
```

```python
# Creació d'histogrames per todas las variables
plt.figure(figsize=(16, 12))

# Iteramos sobre todas las columnas
for i, col in enumerate(df.columns):
    plt.subplot(len(df.columns) // 3 + 1, 3, i+1)
    sns.histplot(df[col], kde=True, color='teal')
    plt.title(f'Histograma de {col}')

plt.tight_layout()
plt.show()
```

```python
df_corr = df.corr(method='spearman')

plt.figure(figsize=(12, 8))

g = sns.heatmap(
    df_corr, 
    cmap='rainbow', 
    annot=True,
    fmt=".2f",  # Formato para mostrar 2 decimales
    center=0,      # Centra la paleta de colors al voltant de 0
    linewidths=0.8 # Separador de cel·les
)

# Afegir títol
plt.title('Matriu de Correlació de les variables', fontsize=18);
```
